```python
# predict_cv2_colab.py

import os
import glob
import cv2
import numpy as np
from tqdm import tqdm
from ultralytics import YOLO

# Extensiones de imagen soportadas
IMG_EXTS = {".jpg", ".jpeg", ".png", ".bmp", ".webp", ".tif", ".tiff"}


# ---------- Iterador de rutas ----------
def iter_paths(src, recursive=False):
    """Itera rutas de imagen en carpeta o archivo √∫nico."""
    if os.path.isdir(src):
        pattern = "**/*" if recursive else "*"
        for ext in IMG_EXTS:
            yield from glob.iglob(os.path.join(src, f"{pattern}{ext}"), recursive=recursive)
            yield from glob.iglob(os.path.join(src, f"{pattern}{ext.upper()}"), recursive=recursive)
    else:
        yield src


# ---------- Recorte de matr√≠culas ----------
def extract_licence_plates(img, result, save_dir):
    """Guarda recortes de las detecciones de matr√≠culas."""
    os.makedirs(save_dir, exist_ok=True)
    if result is None or result.boxes is None or len(result.boxes) == 0:
        return 0

    h, w = img.shape[:2]
    count = 0
    boxes = result.boxes

    for i in range(len(boxes)):
        x1, y1, x2, y2 = boxes.xyxy[i].cpu().numpy().astype(int).tolist()
        x1 = max(0, min(x1, w - 1))
        x2 = max(0, min(x2, w))
        y1 = max(0, min(y1, h - 1))
        y2 = max(0, min(y2, h))

        if x2 <= x1 or y2 <= y1:
            continue

        crop = img[y1:y2, x1:x2]
        if crop.size == 0:
            continue

        out_name = f"plate_{count:02d}.jpg"
        out_path = os.path.join(save_dir, out_name)
        cv2.imwrite(out_path, crop)
        count += 1

    return count


# ---------- MAIN ----------
def main():
    # ===== CONFIGURACI√ìN =====
    # Cambia estas rutas seg√∫n tu Drive
    BASE_PATH = "/content/gdrive/MyDrive/Licence_plate_colab"
    MODEL_PATH = os.path.join(BASE_PATH, "best.pt")        # tu modelo YOLO
    SOURCE_PATH = os.path.join(BASE_PATH, "provaaa.jpg")       # carpeta de im√°genes
    OUT_DIR = os.path.join(BASE_PATH, "yolo_outputs")      # carpeta de salida
    CONF_THRES = 0.5
    IMG_SIZE = 1280

    os.makedirs(OUT_DIR, exist_ok=True)
    PLATES_DIR = os.path.join(OUT_DIR, "plates")
    os.makedirs(PLATES_DIR, exist_ok=True)

    print(f"[INFO] Guardando resultados en: {OUT_DIR}")

    # Detectar GPU autom√°ticamente
    try:
        import torch
        device = "cuda" if torch.cuda.is_available() else "cpu"
    except:
        device = "cpu"
    print(f"[INFO] Usando dispositivo: {device}")

    # Cargar modelo YOLO
    model = YOLO(MODEL_PATH)

    paths = list(iter_paths(SOURCE_PATH))
    print(f"[INFO] Im√°genes encontradas: {len(paths)}")

    processed = 0
    for path in tqdm(paths, desc="Procesando im√°genes", unit="img"):
        img = cv2.imread(path)
        if img is None:
            print(f"[WARN] No pude leer: {path}")
            continue

        results = model.predict(
            source=img,
            imgsz=IMG_SIZE,
            conf=CONF_THRES,
            device=device,
            verbose=False
        )

        if not results or results[0] is None:
            print(f"[WARN] Sin resultados en: {path}")
            continue

        res0 = results[0]
        annotated = res0.plot()  # BGR
        base = os.path.splitext(os.path.basename(path))[0]
        out_path = os.path.join(OUT_DIR, base + "_pred.jpg")
        cv2.imwrite(out_path, annotated)

        per_image_dir = os.path.join(PLATES_DIR, base)
        saved = extract_licence_plates(img, res0, per_image_dir)
        if saved > 0:
            print(f"üì¶ {saved} matr√≠cula(s) guardada(s) ‚Üí {per_image_dir}")
        else:
            print("‚Ñπ No se detectaron matr√≠culas para recortar.")
        processed += 1

    print(f"[DONE] Im√°genes procesadas: {processed}")


if __name__ == "__main__":
    main()
```

```python
# dataset_generator_colab.py
import os
import cv2
import glob
import numpy as np
from tqdm import tqdm

# -----------------------------
# üß© FUNCIONES AUXILIARES
# -----------------------------

import shutil

def limpiar_directorio(dir_path):
    if os.path.exists(dir_path):
        print(f"üßπ Limpiando {dir_path}...")
        shutil.rmtree(dir_path)
    os.makedirs(dir_path, exist_ok=True)

def augment_char(img):
    rows, cols = img.shape
    # Rotaci√≥n leve
    ang = np.random.uniform(-8, 8)
    M = cv2.getRotationMatrix2D((cols / 2, rows / 2), ang, 1)
    img = cv2.warpAffine(img, M, (cols, rows), borderValue=0)
    # Brillo / contraste
    alpha = np.random.uniform(0.8, 1.3)
    beta = np.random.uniform(-30, 30)
    img = cv2.convertScaleAbs(img, alpha=alpha, beta=beta)
    # Blur o ruido
    if np.random.rand() < 0.3:
        img = cv2.GaussianBlur(img, (3, 3), 0)
    if np.random.rand() < 0.2:
        noise = np.random.randint(0, 40, img.shape, dtype=np.uint8)
        img = cv2.add(img, noise)
    return img


def generar_dataset_sintetico(chars, out_dir, n_por_clase=1000):
    os.makedirs(out_dir, exist_ok=True)
    for ch in tqdm(chars, desc="Generando clases", unit="char"):
        ch_dir = os.path.join(out_dir, ch)
        os.makedirs(ch_dir, exist_ok=True)
        for i in range(n_por_clase):
            img = np.ones((64, 44), np.uint8) * 255  # fondo blanco
            cv2.putText(
                img, ch, (5, 55), cv2.FONT_HERSHEY_SIMPLEX,
                1.7, 0, 2, cv2.LINE_AA  # texto negro
            )
            img = augment_char(img)
            cv2.imwrite(os.path.join(ch_dir, f"{ch}_{i:04d}.png"), img)
    print(f"[‚úÖ] Dataset sint√©tico creado en {out_dir}")


def generar_dataset_manual(placas_dir, out_dir):
    """
    Versi√≥n de ejemplo ‚Äî requiere funciones del script KNN_OCR.
    En Colab, solo recomendable si se ejecuta localmente (requiere ventana cv2).
    """
    from KNN_OCR import preprocesar_placa, _segmentar, top7_por_altura, normalizar_char

    os.makedirs(out_dir, exist_ok=True)
    for path in glob.glob(os.path.join(placas_dir, "*.jpg")):
        img = cv2.imread(path)
        if img is None:
            continue
        _, bw = preprocesar_placa(img)
        boxes = _segmentar(bw, "auto")
        boxes = top7_por_altura(boxes)

        for i, box in enumerate(boxes):
            ch_img = normalizar_char(img, box)
            ch_img = cv2.resize(ch_img, (44, 64))
            cv2.imshow("Char", ch_img)
            print("‚Üí Teclea la letra/n√∫mero o ESC para saltar")

            while True:
                key = cv2.waitKey(0)
                if key == 27:  # ESC
                    ch = None
                    break
                if 32 <= key <= 126:
                    ch = chr(key).upper()
                    if ch.isalnum():
                        break
                print("Tecla no v√°lida, prueba otra vez...")

            if ch is None:
                continue

            class_dir = os.path.join(out_dir, ch)
            os.makedirs(class_dir, exist_ok=True)
            n = len(os.listdir(class_dir))
            cv2.imwrite(os.path.join(class_dir, f"{ch}_{n:04d}.png"), ch_img)

    cv2.destroyAllWindows()
    print(f"[OK] Dataset manual guardado en {out_dir}")


# -----------------------------
# üöÄ MAIN AUTOM√ÅTICO (COLAB)
# -----------------------------
if __name__ == "__main__":
    print("=== GENERADOR DE DATASET OCR (COLAB) ===")

    # ===== CONFIGURACI√ìN =====
    BASE_PATH = "/content/gdrive/MyDrive/Licence_plate_colab"
    OUT_NUMEROS = os.path.join(BASE_PATH, "dataset_chars/numeros")
    OUT_LETRAS = os.path.join(BASE_PATH, "dataset_chars/letras")

    DIGITOS = "0123456789"
    LETRAS = "BCDFGHJKLMNPQRSTVWXYZ"

        # Limpiar datasets existentes
    limpiar_directorio(OUT_NUMEROS)
    limpiar_directorio(OUT_LETRAS)


    # ===== GENERAR DATOS =====
    print("üî† Generando dataset de n√∫meros...")
    generar_dataset_sintetico(DIGITOS, OUT_NUMEROS, n_por_clase=2000)

    print("üî§ Generando dataset de letras...")
    generar_dataset_sintetico(LETRAS, OUT_LETRAS, n_por_clase=2000)

    print("[‚úÖ] Dataset sint√©tico completo en Google Drive.")
```

```python
 #Morpholical derivates
def BlackHat(img, kernel=cv2.getStructuringElement(cv2.MORPH_RECT, (15, 5))):
    # Apply the black hat transform
    return cv2.morphologyEx(img.astype(np.uint8), cv2.MORPH_BLACKHAT, kernel, iterations=3)

def Median_filter(img, ks=7):
    return cv2.medianBlur(img.astype(np.uint8),ks)

def GaussianBlur(img):
    return cv2.GaussianBlur(img, (7,7), 0)

#Morphologicall
def Close(img, kernel=rectKernel):
    return cv2.morphologyEx(img, cv2.MORPH_CLOSE, kernel,iterations = 2)

def Dilate(img, kernel=np.ones((5,15), np.uint8), iterations=5):
    return cv2.dilate(img,kernel,iterations = iterations)

def Open(img, kernel=np.ones((3,3), np.uint8)):
    return cv2.morphologyEx(img, cv2.MORPH_OPEN, kernel,iterations = 2)


def Erode(img, kernel=np.ones((3,3), np.uint8), iterations=2):
    return cv2.erode(img,kernel,iterations = iterations)


def Sobel_X(img):
    return cv2.Sobel(img, cv2.CV_32F, dx=1,dy=0, ksize=-1)

def Sobel_Y(img):
    return cv2.Sobel(img, cv2.CV_8U, 0, 1)

#min_max_scaler
def normalization(img):
    img = np.absolute(img)
    (minVal, maxVal) = (np.min(img), np.max(img))
    img = (255 * ((img - minVal) / (maxVal - minVal))).astype("uint8")
    return img

def substracting(img1, img2):
    return cv2.subtract(img1,img2)

def binarization(img):
    #th, im_th = cv2.threshold(img, 128, img.max(), cv2.THRESH_BINARY)
    return cv2.threshold(img, .4*np.max(img), 255,cv2.THRESH_BINARY )[1]#substracted = substracting(dilated, image)
```

```python
def filt_contourns(img):
    cnts = cv2.findContours(img.copy(), cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_SIMPLE)
    cnts = imutils.grab_contours(cnts)
    #cnts = sorted(cnts, key=cv2.contourArea, reverse=True)
    # return the list of contours
    return cnts


def locate_license_plate(gray, candidates,clearBorder=False):
    # initialize the license plate contour and ROI
    lpCnt = None
    roi = None
    ar_list = []
    maxi = 0
    # loop over the license plate candidate contours
    for c in candidates:
        # compute the bounding box of the contour and then use
        # the bounding box to derive the aspect ratio
        (x, y, w, h) = cv2.boundingRect(c)
        area = cv2.contourArea(c)
        ar = w / float(h)
        ar_list.append((ar, len(c), area))
        if ar > 2.5 and ar < 7:
            if area > maxi:
                maxi = copy.copy(area)
                # store the license plate contour and extract the
                # license plate from the grayscale image and then
                # threshold it
                lpCnt = c
                licensePlate = gray[y-25:y + h, x:x + w]
                roi = cv2.threshold(licensePlate, 0, 255,cv2.THRESH_BINARY_INV | cv2.THRESH_OTSU)[1]


    return roi, lpCnt, ar_list
```

```python
import random
from skimage.transform import rotate
def vertical_flip(image):
    return cv2.flip(image, 0)

# NO t√© sentit
def horizontal_flip(image):
    return cv2.flip(image, 1)

def image_dilate(img, kernel=np.ones((3,3), np.uint8), iterations=1):
    return cv2.dilate(img,kernel,iterations = iterations)

def image_translation(image):
    inverse = image.max() - image
    shift_X = random.randint(-10,10)
    shift_Y = random.randint(-10,10)
    M = np.float32([[1, 0, shift_X], [0, 1, shift_Y]])
    translated = cv2.warpAffine(inverse, M, (inverse.shape[1], inverse.shape[0]))
    return translated.max() - translated

def image_rotation(image):
    inverse = image.max() - image
    degrees = random.randint(-40,40)
    (h, w) = inverse.shape[:2]
    (cX, cY) = (w // 2, h // 2)
    M = cv2.getRotationMatrix2D((cX, cY), degrees, 1.0)
    result = cv2.warpAffine(inverse, M, (w, h))
    return result.max() - result

def image_crop(image):
    scale = 0.8+0.2*random.random()
    height, width = int(image.shape[0]*scale), int(image.shape[1]*scale)
    x = random.randint(0, image.shape[1] - int(width))
    y = random.randint(0, image.shape[0] - int(height))
    cropped = image[y:y+height, x:x+width]
    return cv2.resize(cropped, (image.shape[1], image.shape[0]))

```

```python
def apply_random_seq(image):
    funcs = [image_rotation, image_translation, image_crop, image_dilate]
    num = random.randint(1,3)
    funcs = random.sample(funcs, num)
    for func in funcs:
        image = func(image.astype(np.uint8))
    return image

#'''
letter_list = [cv2.resize(read(gt_folder_letters + x), SHAPE) != 0 for x in os.listdir(gt_folder_letters)]
number_list = [cv2.resize(read(gt_folder_numbers + x), SHAPE) != 0 for x in os.listdir(gt_folder_numbers)]

for x, letter in zip(letter_list,os.listdir(gt_folder_letters)):
    for count in range(500):
        modified = apply_random_seq(x)
        cv2.imwrite(gt_folder_letters+letter[0]+'.'+str(count)+'.png', modified*255)

for x, number in zip(number_list,os.listdir(gt_folder_numbers)):
    for count in range(500):
        modified = apply_random_seq(x)
        cv2.imwrite(gt_folder_numbers+number[0]+'.'+str(count)+'.png', modified*255)
#'''
```

```python
import numpy as np
import cv2
import os
from sklearn.neighbors import KNeighborsClassifier

# Shape deseado
SHAPE = (30, 20)

# Funci√≥n de features
def features(img):
    return img.flatten()

# Funci√≥n de augmentation
def augment(img):
    # Rotaci√≥n leve
    rows, cols = img.shape
    ang = np.random.uniform(-8, 8)
    M = cv2.getRotationMatrix2D((cols/2, rows/2), ang, 1)
    img = cv2.warpAffine(img, M, (cols, rows), borderValue=0)

    # Brillo/contraste
    alpha = np.random.uniform(0.8, 1.3)
    beta = np.random.uniform(-30, 30)
    img = cv2.convertScaleAbs(img, alpha=alpha, beta=beta)

    # Blur
    if np.random.rand() < 0.3:
        img = cv2.GaussianBlur(img, (3, 3), 0)

    # Ruido
    if np.random.rand() < 0.2:
        noise = np.random.randint(0, 40, img.shape, dtype=np.uint8)
        img = cv2.add(img, noise)

    return img

# -----------------------
# LETRAS
# -----------------------
letter_list = []
for letra in os.listdir(gt_folder_letters):
    letra_folder = os.path.join(gt_folder_letters, letra)
    if not os.path.isdir(letra_folder):
        continue
    for archivo in os.listdir(letra_folder):
        ruta = os.path.join(letra_folder, archivo)
        img = cv2.imread(ruta, cv2.IMREAD_GRAYSCALE)
        if img is None:
            continue
        img = cv2.resize(img, SHAPE)

        # Data augmentation: crea 3 versiones aumentadas por ejemplo
        for _ in range(3):
            img_aug = augment(img)
            letter_list.append((features(img_aug != 0), letra))

X_letters = [x[0] for x in letter_list]
y_letters = [x[1] for x in letter_list]

# -----------------------
# N√öMEROS
# -----------------------
number_list = []
for number in os.listdir(gt_folder_numbers):
    number_folder = os.path.join(gt_folder_numbers, number)
    if not os.path.isdir(number_folder):
        continue
    for archivo in os.listdir(number_folder):
        ruta = os.path.join(number_folder, archivo)
        img = cv2.imread(ruta, cv2.IMREAD_GRAYSCALE)
        if img is None:
            continue
        img = cv2.resize(img, SHAPE)

        # Data augmentation
        for _ in range(3):
            img_aug = augment(img)
            number_list.append((features(img_aug != 0), number))

X_numbers = [x[0] for x in number_list]
y_numbers = [x[1] for x in number_list]


print("N√∫mero de letras cargadas:", len(letter_list))
print("N√∫mero de n√∫meros cargados:", len(number_list))


# -----------------------
# Entrenamiento KNN
# -----------------------
letters_classifier_KNN = KNeighborsClassifier(n_neighbors=10)
letters_classifier_KNN.fit(X_letters, y_letters)

numbers_classifier_KNN = KNeighborsClassifier(n_neighbors=10)
numbers_classifier_KNN.fit(X_numbers, y_numbers)
```

```python
import cv2
import numpy as np
import copy
import random
from collections import Counter
from skimage import measure
import imutils

# -----------------------------
# Funciones de augmentaci√≥n
# -----------------------------
def vertical_flip(image):
    return cv2.flip(image.astype(np.uint8), 0)

def horizontal_flip(image):
    return cv2.flip(image.astype(np.uint8), 1)


def image_dilate(img, kernel=np.ones((3,3), np.uint8), iterations=1):
    return cv2.dilate(img, kernel, iterations=iterations)

def image_translation(image):
    inverse = 255 - image.astype(np.uint8)
    shift_X = random.randint(-10,10)
    shift_Y = random.randint(-10,10)
    M = np.float32([[1, 0, shift_X], [0, 1, shift_Y]])
    translated = cv2.warpAffine(inverse, M, (inverse.shape[1], inverse.shape[0]))
    return 255 - translated


def image_rotation(image):
    inverse = 255 - image.astype(np.uint8)
    degrees = random.randint(-40,40)
    (h, w) = inverse.shape[:2]
    (cX, cY) = (w // 2, h // 2)
    M = cv2.getRotationMatrix2D((cX, cY), degrees, 1.0)
    result = cv2.warpAffine(inverse, M, (w, h))
    return 255 - result


def image_crop(image):
    scale = 0.8 + 0.2 * random.random()
    height, width = int(image.shape[0]*scale), int(image.shape[1]*scale)
    x = random.randint(0, image.shape[1] - width)
    y = random.randint(0, image.shape[0] - height)
    cropped = image[y:y+height, x:x+width]
    return cv2.resize(cropped, (image.shape[1], image.shape[0]))

def augment_crop(crop):
    crop = crop.astype(np.uint8)  # ‚ö° asegura que OpenCV funcione
    funcs = [vertical_flip, horizontal_flip, image_dilate,
             image_translation, image_rotation, image_crop]
    aug = crop.copy()
    for _ in range(random.randint(1,2)):
        func = random.choice(funcs)
        aug = func(aug)
    return aug


# -----------------------------
# Funci√≥n de features
# -----------------------------
def features(img):
    return img.flatten()

# -----------------------------
# Funci√≥n de predicci√≥n
# -----------------------------
def predict_image_with_augmentation(filename, letters_classifier, numbers_classifier, SHAPE=(30,20), n_aug=5):
    base = '/content/gdrive/MyDrive/Licence_plate_colab/'
    img_base = read(base + filename)  # tu funci√≥n read() que devuelve RGB o grayscale

    # ---------------------
    # Preprocesamiento
    # ---------------------
    img = BlackHat(img_base)
    img = Sobel_X(img)
    img = normalization(img)
    img = GaussianBlur(img)
    img = Close(img)
    img = binarization(img)
    img = Median_filter(img, 9)
    img = Dilate(img)
    img = Close(img, cv2.getStructuringElement(cv2.MORPH_RECT, (20, 30)))

    contourns = filt_contourns(img)
    final, _, ar_list = locate_license_plate(img_base, contourns)
    final = copy.copy(Erode(final, iterations=1))

    labels = measure.label(final, connectivity=2, background=0)
    charCandidates = np.zeros(final.shape, dtype="uint8")

    for label in np.unique(labels):
        if label == 0:
            continue
        labelMask = np.zeros(final.shape, dtype="uint8")
        labelMask[labels == label] = 255
        cnts = cv2.findContours(labelMask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
        cnts = imutils.grab_contours(cnts)
        if len(cnts) > 0:
            c = max(cnts, key=cv2.contourArea)
            (boxX, boxY, boxW, boxH) = cv2.boundingRect(c)
            aspectRatio = boxW / float(boxH)
            solidity = cv2.contourArea(c) / float(boxW * boxH)
            heightRatio = boxH / float(final.shape[0])
            if aspectRatio < 1.0 and solidity > 0.15 and 0.4 < heightRatio < 0.95:
                hull = cv2.convexHull(c)
                cv2.drawContours(charCandidates, [hull], -1, 255, -1)

    countors = filt_contourns(charCandidates)
    countors = sorted(countors, key=lambda x: x[0][0][0], reverse=True)
    crops = [final[y:y+h, x:x+w] for img in countors for x,y,w,h in [cv2.boundingRect(img)]]
    crops = clean_crops(crops)

    # ---------------------
    # Predicci√≥n con augmentaci√≥n
    # ---------------------
    def predict_crop_list(crop_list, classifier):
        predictions = []
        for crop in crop_list:
            aug_crops = [augment_crop(crop) for _ in range(n_aug)]
            aug_features = [features(cv2.resize((x != 0).astype(np.uint8), SHAPE)) for x in aug_crops]
            preds = classifier.predict(aug_features)
            most_common = Counter(preds).most_common(1)[0][0]
            predictions.append(most_common)
        return predictions

    letters = predict_crop_list(crops[4:], letters_classifier)
    numbers = predict_crop_list(crops[:4], numbers_classifier)

    return ''.join(numbers + letters)
```

```python
import cv2
import os
import numpy as np

def save_all_crops(crops, out_dir):
    """
    Guarda todos los crops en out_dir con nombres √∫nicos.
    """
    os.makedirs(out_dir, exist_ok=True)
    counter = 0

    for crop in crops:
        # Normalizar a uint8
        if crop.dtype == np.bool_:
            crop_to_save = (crop.astype(np.uint8)) * 255
        elif crop.dtype in [np.float32, np.float64]:
            crop_to_save = (np.clip(crop, 0, 1) * 255).astype(np.uint8)
        else:
            crop_to_save = np.clip(crop, 0, 255).astype(np.uint8)

        filename = f"crop_{counter:04d}.png"
        path = os.path.join(out_dir, filename)
        cv2.imwrite(path, crop_to_save)
        counter += 1

    print(f"[‚úÖ] Guardadas {counter} im√°genes en {out_dir}")


# Carpeta de salida
out_dir = "/content/gdrive/MyDrive/Licence_plate_colab/dataset_chars/numeros"

# Guardar todos los crops
save_all_crops(crops, out_dir)
```
